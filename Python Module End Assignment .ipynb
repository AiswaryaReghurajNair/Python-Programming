```python
#Preprocessing
abc_data['Height']=np.random.randint(150,181, size=abc_data.shape[0])
abc_data[['Name','Height']].head()
```

```python
abc_data['College']=abc_data['College'].fillna('Unknown')
abc_data[['Name','College']].head()
```

```python
abc_data.isnull().sum()
```

```python
abc_data['Salary']=abc_data['Salary'].fillna(abc_data['Salary'].mean)
```

```python
abc_data['Salary'].isnull().sum()
```

```python
print(abc_data['Salary'].head())
```

```python
#1
team_distribution=abc_data['Team'].value_counts()
team_percentage_split=(team_distribution/abc_data.shape[0]*100)
team_distribution_df=pd.DataFrame({
    'Employee count':team_distribution,
    'Percentage Split':team_percentage_split
})
team_distribution_df

plt.figure(figsize=(12,6))
team_distribution_df.plot(kind='bar', color='green')
plt.title("Percnetage of Employees in each Team")
plt.xlabel("Team")
plt.ylabel("Employee Count")
plt.show()
```

```python
#2
position_distribution=abc_data['Position'].value_counts()
print(position_distribution)

plt.figure(figsize=(10,6))
position_distribution.plot(kind='bar', color='blue')
plt.title("No. of employees related to position")
plt.xlabel('Position')
plt.ylabel("No. of employees")
plt.show()
```

```python
#3
age_bins=[18,25,35,45,60]
age_labels=['18-25','26-35','36-45','46-60']
abc_data['Age Group']=pd.cut(abc_data['Age'], bins=age_bins, labels=age_labels)
age_group_distribution=abc_data['Age Group'].value_counts().sort_index()
plt.figure(figsize=(10,6))
age_group_distribution.plot(kind='bar', color='blue')
plt.title("Age group")
plt.xlabel('Age Group')
plt.ylabel("No. of employees")
plt.show()
```

```python
#Preprocessing
abc_data['Salary'] = pd.to_numeric(abc_data['Salary'], errors='coerce')
print(abc_data['Salary'].isna().sum())
print(f"NaN values in Salary: {abc_data['Salary'].isna().sum()}")
abc_data.dropna(subset=['Salary'], inplace=True)
#4
team_salary_expenditure=abc_data.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary_expenditure=abc_data.groupby('Position')['Salary'].sum().sort_values(ascending=False)
#plot
plt.figure(figsize=(10,6))
team_salary_expenditure.plot(kind='bar',color='red')
plt.title("Salary expenditure based on Team")
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
position_salary_expenditure.plot(kind='bar', color='orange')
plt.title("Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Total Salary Expenditure")
plt.show()
```

```python
#5
plt.figure(figsize=(8, 6))
sb.scatterplot(x='Age', y='Salary', data=abc_data, color='blue')
plt.title("Correlation between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
